```python
from PIL import Image

# Fayl yolunun əsas hissəsi
base_path = r'C:\Users\dell\Desktop\Power BI\Fusion Academy\Pictures\Ders1_elave_'

# Şəkilləri saxlayacağımız siyahı
image_list = []

# 1-dən 5-ə qədər olan şəkilləri dövr ilə açırıq
for i in range(1, 3):  # 1-dən 5-ə qədər
    full_path = f'{base_path}{i}.jpg'  # Fayl yolunu dinamik olaraq yaradırıq
    img = Image.open(full_path)
    image_list.append(img)  # Şəkili siyahıya əlavə edirik

# Nəticə: images siyahısında bütün yüklənmiş şəkillər olacaq

import matplotlib.pyplot as plt

# Şəkilləri göstərin
for i, image in enumerate(image_list):
    if i == 0:
        plt.figure(figsize=(10, 5))  # Şəkilin ölçüsü
        plt.title(" 1-ci  Sətiri Başlıq və ya Başlığı Sətir etmək", fontsize=18, fontweight='bold')
    if i == 1:
        plt.figure(figsize=(10, 5))  # Şəkilin ölçüsü
        plt.title("""Null silinməsi (Remove Blank Rows)-: Qəyd hər Sətirin bütün 
                    dəyərləri Null olmalıdır""", fontsize=18, fontweight='bold')        
    plt.imshow(image)        
    plt.imshow(image)
    plt.axis('off')  # Oxları gizlətmək üçün
    plt.show()
```
