```python
from PIL import Image

# Fayl yolunun əsas hissəsi
base_path = r'C:\Users\dell\Desktop\Power BI\Fusion Academy\Pictures\Ders1_'

# Şəkilləri saxlayacağımız siyahı
image_list = []

# 1-dən 5-ə qədər olan şəkilləri dövr ilə açırıq
for i in range(1, 15):  # 1-dən 5-ə qədər
    full_path = f'{base_path}{i}.jpg'  # Fayl yolunu dinamik olaraq yaradırıq
    img = Image.open(full_path)
    image_list.append(img)  # Şəkili siyahıya əlavə edirik

# Nəticə: images siyahısında bütün yüklənmiş şəkillər olacaq

import matplotlib.pyplot as plt

# Şəkilləri göstərin
for i, image in enumerate(image_list):
    if i == 0:
        plt.figure(figsize=(10, 5))  # Şəkilin ölçüsü
        plt.title("Datanın tipini dəyişmək", fontsize=18, fontweight='bold')
    elif i == 1:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Data növünü seç", fontsize=18, fontweight='bold')
    elif i == 2:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Əməliyyatı geri qaytarmaq", fontsize=18, fontweight='bold')
    elif i == 3:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Dəyər dəyişdirmək", fontsize=18, fontweight='bold')
    elif i == 5:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Datanı filterləmək", fontsize=18, fontweight='bold')
    elif i == 7:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Yeni Column yaratmaq (Column From Examples)", fontsize=18, fontweight='bold')
    elif i == 8:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Yeni Column yaratmaq (Custom Column)", fontsize=18, fontweight='bold')
    elif i == 9:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Şərt ilə Column yaratmaq (Conditional Column)", fontsize=18, fontweight='bold')
    elif i == 10:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Sutunu ayırmaq (Split Column)", fontsize=18, fontweight='bold')
    elif i == 11:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Cədvəlləri alt-alta birləşdirmək (Append Queries)", fontsize=18, fontweight='bold')
    elif i == 12:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Cədvəlləri yan-yan birləşdirmək (Merge)", fontsize=18, fontweight='bold')
    elif i == 13:
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        plt.title("Unpivot table", fontsize=18, fontweight='bold')        
    plt.imshow(image)        
    plt.imshow(image)
    plt.axis('off')  # Oxları gizlətmək üçün
    plt.show()
```
