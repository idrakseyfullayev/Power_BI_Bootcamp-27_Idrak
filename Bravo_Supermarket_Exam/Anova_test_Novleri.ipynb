## ANOVA n√∂vl…ôri



**ANOVA** bir ne√ß…ô n√∂v…ô b√∂l√ºn√ºr. **One-way ANOVA** onlarƒ±n sad…ô formasƒ±dƒ±r, amma ba≈üqa n√∂vl…ôri d…ô var:


### üîπ **ANOVA n√∂vl…ôri**

1Ô∏è‚É£ **One-Way ANOVA**

* Bir m√ºst…ôqil d…ôyi≈ü…ôn (faktor) ‚Üí √ßox qrupun ortalamalarƒ±nƒ± m√ºqayis…ô edir.
* M…ôs…ôl…ôn: ‚ÄúF…ôrqli filiallarda maa≈ü ortalamalarƒ± eyni deyilmi?‚Äù

---

2Ô∏è‚É£ **Two-Way ANOVA**

* ƒ∞ki faktor n…ôz…ôr…ô alƒ±nƒ±r (v…ô onlarƒ±n qar≈üƒ±lƒ±qlƒ± t…ôsiri d…ô √∂l√ß√ºl…ô bil…ôr).
* M…ôs…ôl…ôn: ‚ÄúMaa≈ü ortalamalarƒ± **cins…ô** v…ô **t…ôhsil…ô** g√∂r…ô d…ôyi≈üirmi? + ƒ∞kisi birlikd…ô t…ôsir g√∂st…ôrirmi?‚Äù

---

3Ô∏è‚É£ **N-way ANOVA (Factorial ANOVA)**

* ƒ∞ki v…ô daha √ßox faktor olduqda.
* √úmumi halda $k$ faktor ola bil…ôr.

---

4Ô∏è‚É£ **Repeated Measures ANOVA**

* Eyni subyektl…ôr √ºz…ôrind…ô t…ôkrar √∂l√ßm…ôl…ôr zamanƒ± istifad…ô olunur.
* M…ôs…ôl…ôn: Eyni t…ôl…ôb…ôl…ôrin **imtahan n…ôtic…ôl…ôri** (1-ci, 2-ci, 3-c√º imtahan).

---

5Ô∏è‚É£ **MANOVA (Multivariate ANOVA)**

* Bird…ôn √ßox dependent (target) d…ôyi≈ü…ôn varsa.
* M…ôs…ôl…ôn: T…ôhsil s…ôviyy…ôsinin h…ôm **maa≈üa**, h…ôm d…ô **i≈ü m…ômnuniyy…ôtin…ô** t…ôsiri.

---

6Ô∏è‚É£ **ANCOVA (Analysis of Covariance)**

* ANOVA + Covariate (kontrol d…ôyi≈ü…ônl…ôr).
* Y…ôni qruplar arasƒ±ndakƒ± f…ôrql…ôri m√ºqayis…ô ed…ôrk…ôn …ôlav…ô d…ôyi≈ü…ônl…ôri n…ôz…ôr…ô alƒ±r.
* M…ôs…ôl…ôn: M√ºxt…ôlif d…ôrs metodlarƒ±nƒ±n t…ôsiri m√ºqayis…ô olunur, amma **t…ôl…ôb…ônin …ôvv…ôlki bilik s…ôviyy…ôsi** d…ô (covariate) n…ôz…ôr…ô alƒ±nƒ±r.

---

üìå Qƒ±sa:

* Sad…ô: **One-Way ANOVA**
* ƒ∞ki faktor: **Two-Way ANOVA**
* √áox faktor: **N-Way ANOVA**
* T…ôkrar √∂l√ßm…ô: **Repeated Measures ANOVA**
* √áoxlu dependent: **MANOVA**
* Covariate il…ô: **ANCOVA**



## üîπ **ANOVA n√∂vl…ôri v…ô Python n√ºmun…ôl…ôri**



### 1Ô∏è‚É£ **One-Way ANOVA**

**M…ôqs…ôd:** Bir faktor √ºzr…ô (bir m√ºst…ôqil d…ôyi≈ü…ôn) √ßox qrupun ortalamalarƒ±nƒ± m√ºqayis…ô etm…ôk.

**Python kodu:**

```python
from scipy import stats

# M…ôs…ôl…ôn: 3 qrup t…ôl…ôb…ô balƒ±
group1 = [85, 90, 88, 75, 95]
group2 = [70, 65, 80, 72, 68]
group3 = [92, 88, 91, 85, 90]

f_stat, p_value = stats.f_oneway(group1, group2, group3)
print("F-stat:", f_stat, "p-value:", p_value)
```

---

### 2Ô∏è‚É£ **Two-Way ANOVA**

**M…ôqs…ôd:** ƒ∞ki faktor v…ô onlarƒ±n qar≈üƒ±lƒ±qlƒ± t…ôsirini yoxlamaq.

**Python kodu (statsmodels il…ô):**

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Sad…ô dataset
df = pd.DataFrame({
    "Score": [85,90,88,75,95, 70,65,80,72,68, 92,88,91,85,90],
    "Gender": ["M","M","M","F","F", "M","M","F","F","M", "F","F","M","M","F"],
    "Method": ["A","B","A","B","A", "B","A","B","A","B", "A","A","B","B","A"]
})

# Two-way ANOVA
model = ols("Score ~ C(Gender) + C(Method) + C(Gender):C(Method)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

---

### 3Ô∏è‚É£ **N-way ANOVA (Factorial ANOVA)**

**M…ôqs…ôd:** ƒ∞ki v…ô daha √ßox faktor (m…ôs…ôl…ôn, 3 faktor).

**Python kodu:**

```python
# √ú√ß faktor n√ºmun…ôsi
model = ols("Score ~ C(Gender) * C(Method) * C(AgeGroup)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

---

### 4Ô∏è‚É£ **Repeated Measures ANOVA**

**M…ôqs…ôd:** Eyni i≈ütirak√ßƒ±larƒ±n t…ôkrar √∂l√ßm…ôl…ôri.

**Python kodu:**

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.anova import AnovaRM

# Dataset: eyni t…ôl…ôb…ôl…ôrin 3 imtahan n…ôtic…ôsi
df = pd.DataFrame({
    "Subject": [1,1,1,2,2,2,3,3,3],
    "Test": ["T1","T2","T3","T1","T2","T3","T1","T2","T3"],
    "Score": [80,85,90,70,75,78,88,92,95]
})

# Repeated Measures ANOVA
anova = AnovaRM(df, "Score", "Subject", within=["Test"]).fit()
print(anova)
```

---

### 5Ô∏è‚É£ **MANOVA (Multivariate ANOVA)**

**M…ôqs…ôd:** Bird…ôn √ßox dependent (target) d…ôyi≈ü…ôn.

**Python kodu:**

```python
from statsmodels.multivariate.manova import MANOVA

df = pd.DataFrame({
    "Math": [85,70,92,88,75],
    "Science": [90,65,88,80,72],
    "Method": ["A","B","A","B","A"]
})

# MANOVA
manova = MANOVA.from_formula("Math + Science ~ Method", data=df)
print(manova.mv_test())
```

---

### 6Ô∏è‚É£ **ANCOVA (Analysis of Covariance)**

**M…ôqs…ôd:** ANOVA + covariate (y…ôni …ôlav…ô kontrol d…ôyi≈ü…ôn).

**Python kodu:**

```python
# ANCOVA n√ºmun…ôsi
df = pd.DataFrame({
    "Score": [85,90,70,65,92,88],
    "Method": ["A","A","B","B","A","B"],
    "PreTest": [78,82,60,58,80,62]  # covariate
})

model = ols("Score ~ C(Method) + PreTest", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

---

* **Repeated Measures** ‚Üí t…ôkrar √∂l√ß√ºl…ôr
* **MANOVA** ‚Üí √ßoxlu dependent d…ôyi≈ü…ôn
