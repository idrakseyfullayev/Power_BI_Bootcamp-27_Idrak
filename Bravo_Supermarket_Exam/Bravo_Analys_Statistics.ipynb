### HR datasetidirsə (işçilərin məlumatı), statistik testləri adətən 3 istiqamətdə aparırlar



## **1️⃣ Kəmiyyət – Kəmiyyət (Korelasiya baxışı)**

**Məqsəd:** Hansı kəmiyyət dəyərlərinin bir-birinə bağlı olduğunu öyrənmək.
Bu, əsasən **Pearson/Spearman korelyasiyası** ilə edilir.

**Vacib cütlər:**

* (Age, Salary) → Yaş artdıqca maaş artırmı?
* (Age, TenureYears) → Yaşlı işçilərin şirkətdə qalma müddəti.
* (Salary, TenureYears) → Uzunmüddətli işçilərin maaş səviyyəsi.
* (PerformanceRating, Salary) → Yüksək performans maaşa təsir edir mi?
* (PerformanceScore, Salary) → Performans skorları maaşla bağlıdırmı?
* (TenureYears, AbsenceDays) → Şirkətdə çox qalanların işdən qalma günləri.
* (TrainingHours, PerformanceScore) → Təlim saatlarının performansa təsiri.

---

## **2️⃣ Kəmiyyət – Kateqoriya (T-test, ANOVA)**

**Məqsəd:** Kəmiyyət dəyişənlərin müxtəlif qruplar üzrə fərqlənib-fərqlənmədiyini yoxlamaq.

**Vacib cütlər:**

* (Salary, Department) → Departamentlər arası maaş fərqi.
* (Salary, Education) → Təhsil səviyyəsinin maaşa təsiri.
* (Salary, Gender) → Gender maaş fərqliliyi.
* (Salary, Region) → Regionlar arası maaş fərqi.
* (PerformanceScore, Department) → Departamentlər üzrə performans fərqi.
* (AbsenceDays, Shift) → Növbəyə görə işdən qalma fərqliliyi.
* (TenureYears, StoreSize) → Mağaza ölçüsünün işçilərin qalma müddətinə təsiri.

---

## **3️⃣ Kateqoriya – Kateqoriya (Chi-Square)**

**Məqsəd:** İki kateqoriya dəyişənin bir-birinə bağlı olub-olmadığını yoxlamaq.

**Vacib cütlər:**

* (Department, Region) → Regionlara görə departament paylanması.
* (Gender, Department) → Departamentlərdə gender payı.
* (Gender, Is Terminated) → İşdən çıxma genderlə bağlıdırmı?
* (Region, StoreType) → Regionlara görə mağaza tipləri.
* (LeaveReason, TerminationType) → İşdən çıxma səbəbi ilə çıxma tipi əlaqəsi.
* (OfferAccepted, SourceOfHire) → İş təklifinin qəbul olunması mənbədən asılıdırmı?

---

### (Age, TenureYears) → Yaşlı işçilərin şirkətdə qalma müddəti

```python
df.columns.sort_values()
```

```python
df[['Age', 'TenureYears']].isnull().sum()
```

```python
df_group = df.copy()
df_group = df_group[['Age', 'TenureYears']].dropna()
df_group
```

```python
from scipy import stats

# NaN-ları sil və yalnız iki sütunu götür
age = df_group['Age']
tenure = df_group['TenureYears']

# Pearson korelyasiyası
pearson_corr, pearson_p = stats.pearsonr(age, tenure)
print(f"Pearson r = {pearson_corr:.3f}, p-value = {pearson_p:.3f}")
if pearson_corr > 0 and pearson_p < 0.05:
    print("Pearson: Yaşlı işçilər şirkətdə daha uzun qalır (müsbət əlaqə).")
elif pearson_corr < 0 and pearson_p < 0.05:
    print("Pearson: Yaşlı işçilər şirkətdə daha az qalır (mənfi əlaqə).")
else:
    print("Pearson: Əhəmiyyətli əlaqə yoxdur.")

# Spearman korelyasiyası
spearman_corr, spearman_p = stats.spearmanr(age, tenure)
print(f"Spearman r = {spearman_corr:.3f}, p-value = {spearman_p:.3f}")
if spearman_corr > 0 and spearman_p < 0.05:
    print("Spearman: Yaşlı işçilər şirkətdə daha uzun qalır (müsbət əlaqə).")
elif spearman_corr < 0 and spearman_p < 0.05:
    print("Spearman: Yaşlı işçilər şirkətdə daha az qalır (mənfi əlaqə).")
else:
    print("Spearman: Əhəmiyyətli əlaqə yoxdur.")
```
