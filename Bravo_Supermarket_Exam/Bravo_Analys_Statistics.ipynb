### (Salary, Department) → Departamentlər arası maaş fərqi.

```python
df.columns.sort_values()
```

```python
df[['Department', 'Salary']]
```

```python
df[['Department', 'Salary']].isnull().sum()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols


df_anova = df[['Salary', 'Department']]

# ANOVA modeli qur
model = ols('Salary ~ C(Department)', data=df_anova).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)

# Əlavə izah
f_value = anova_table['F'][0]
p_value = anova_table['PR(>F)'][0]

print(f"\nF-statistic = {f_value:.3f}, p-value = {p_value:.3f}")
if p_value < 0.05:
    print("Departamentlər arasında maaş fərqi əhəmiyyətlidir.")
else:
    print("Departamentlər arasında maaş fərqi əhəmiyyətli deyil.")
```

```python
from scipy import stats


df_anova = df[['Salary', 'Department']]

# Departamentləri qruplara ayır
groups = [group['Salary'].values for name, group in df_anova.groupby('Department')]

# ANOVA F-test
f_stat, p_val = stats.f_oneway(*groups)

print(f"F-statistic = {f_stat:.3f}, p-value = {p_val:.3f}")
if p_val < 0.05:
    print("Departamentlər arasında maaş fərqi əhəmiyyətlidir.")
else:
    print("Departamentlər arasında maaş fərqi əhəmiyyətli deyil.")
```
