### HR datasetidirsə (işçilərin məlumatı), statistik testləri adətən 4 istiqamətdə aparırlar

## **1️⃣ Kəmiyyət – Nəzəri Orta ilə Test (One-Sample t-Test)**

**Məqsəd:** Hər bir kəmiyyət dəyərinin ortalamasının müəyyən bir nəzəri dəyərdən statistik olaraq fərqlənib-fərqlənmədiyini öyrənmək. Bu, əsasən **two-sided one-sample t-test** ilə edilir. Həmçinin, lazım olduqda **one-sided t-test** ilə ortalamanın nəzəri dəyərdən **kiçik və ya böyük** olub-olmadığı da yoxlanıla bilər.

**Vacib nümunələr:**

1. (Salary) → İşçilərin maaş ortalaması müəyyən bir dəyərdən fərqlidirmi?

   * Məsələn: Ortalama maaş 5000-dən fərqlidirmi?
   * One-sided testlər:

     * Ortalama **700-dən kiçikdir**?
     * Ortalama **700-dən böyükdür**?

2. (TenureYears) → İşçilərin şirkətdə qalma müddətinin ortalaması nəzəri dəyərdən fərqlidirmi?

   * Məsələn: Ortalama 2 il dən fərqlidirmi.
   * One-sided testlər:

     * Ortalama **2 ildən kiçikdir**?
     * Ortalama **2 ildən böyükdür**?

---

##  **2️⃣ Kəmiyyət – Kəmiyyət (Korelasiya baxışı)**

**Məqsəd:** Hansı kəmiyyət dəyərlərinin bir-birinə bağlı olduğunu öyrənmək.
Bu, əsasən **Pearson/Spearman korelyasiyası** ilə edilir.

**Vacib cütlər:**

* (Age, Salary) → Yaş artdıqca maaş artırmı?
* (Age, TenureYears) → Yaşlı işçilərin şirkətdə qalma müddəti.
* (Salary, TenureYears) → Uzunmüddətli işçilərin maaş səviyyəsi.
* (PerformanceScore, Salary) → Performans skorları maaşla bağlıdırmı?
* (TenureDays, AbsenceDays) → Şirkətdə çox qalanların işdən qalma günləri.
* (TrainingHours, PerformanceScore) → Təlim saatlarının performansa təsiri.

---

##  3️⃣ Kəmiyyət – Kateqoriya (T-test, ANOVA)**

**Məqsəd:** Kəmiyyət dəyişənlərin müxtəlif qruplar üzrə fərqlənib-fərqlənmədiyini yoxlamaq.

**Vacib cütlər:**

* (Salary, Department) → Departamentlər arası maaş fərqi.
* (Salary, Education) → Təhsil səviyyəsinin maaşa təsiri.
* (Salary, Gender) → Gender maaş fərqliliyi.
* (Salary, Region) → Regionlar arası maaş fərqi.
* (PerformanceScore, Department) → Departamentlər üzrə performans fərqi.
* (PerformanceRating, Salary) → Yüksək performans maaşa təsir edir mi?
* (AbsenceDays, Shift) → Növbəyə görə işdən qalma fərqliliyi.
* (TenureYears, StoreSize) → Mağaza ölçüsünün işçilərin qalma müddətinə təsiri.

---

## **4️⃣ Kateqoriya – Kateqoriya (Chi-Square)**

**Məqsəd:** İki kateqoriya dəyişənin bir-birinə bağlı olub-olmadığını yoxlamaq.

**Vacib cütlər:**

* (Department, Region) → Regionlara görə departament paylanması.
* (Gender, Department) → Departamentlərdə gender payı.
* (Gender, Is Terminated) → İşdən çıxma genderlə bağlıdırmı?

### Cohen's d & Hedges' g & 95% Confidence interval

```python
import pandas as pd
import numpy as np
from scipy import stats

# NaN-ları sil və numeric dəyişənə çevir
salary = df['Salary']

# Null hipotez: ortalama = 700
theoretical_mean = 700
n = len(salary)
sample_mean = np.mean(salary)
sample_std = np.std(salary, ddof=1)  # sample standard deviation

# One-sample t-test
t_stat, p_val = stats.ttest_1samp(salary, theoretical_mean)

# Cohen's d
cohens_d = (sample_mean - theoretical_mean) / sample_std

# Hedges' g (nümunə ölçüsünə düzəliş)
correction = 1 - (3 / (4*n - 5))
hedges_g = cohens_d * correction

# 95% Confidence interval
se = sample_std / np.sqrt(n)
ci_lower = sample_mean - 1.96*se
ci_upper = sample_mean + 1.96*se

# Nəticələri çap et
print(f"Sample mean = {sample_mean:.3f}, Sample std = {sample_std:.3f}, n = {n}")
print(f"T-statistic = {t_stat:.3f}, p-value = {p_val:.3f}")
print(f"Cohen's d = {cohens_d:.3f}")
print(f"Hedges' g = {hedges_g:.3f}")
print(f"95% Confidence interval = ({ci_lower:.3f}, {ci_upper:.3f})")

# Nəticənin şərhi
if p_val < 0.05:
    print("Ortalama maaş nəzəri dəyərdən statistik olaraq fərqlidir.")
else:
    print("Ortalama maaş nəzəri dəyərdən statistik olaraq fərqli deyil.")
```

### Eta-squared (η²) və Omega-squared (ω²) və Tukey

```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def anova_effect_sizes_with_tukey(df, group_col, score_col):
    """
    Bir neçə qrup üçün:
    - ANOVA
    - Eta-squared (η²)
    - Omega-squared (ω²)
    - Tukey HSD post-hoc testi
    """
    # Qrupları ayırırıq
    groups = [df[df[group_col] == g][score_col].values for g in df[group_col].unique()]

    # --- ANOVA test ---
    f_stat, p_val = stats.f_oneway(*groups)
    
    # Bütün datanı birləşdiririk
    all_data = df[score_col].values
    grand_mean = np.mean(all_data)
    N = len(all_data)
    k = len(groups)

    # --- SS_total ---
    SS_total = np.sum((all_data - grand_mean) ** 2)

    # --- SS_between ---
    ns = [len(g) for g in groups]
    means = [np.mean(g) for g in groups]
    SS_between = sum([n * (m - grand_mean) ** 2 for n, m in zip(ns, means)])

    # --- SS_within ---
    SS_within = SS_total - SS_between

    # --- Azadlıq dərəcələri ---
    df_between = k - 1
    df_within = N - k

    # --- MS_within ---
    MS_within = SS_within / df_within

    # --- Effekt ölçüləri ---
    eta_sq = SS_between / SS_total
    omega_sq = (SS_between - (df_between * MS_within)) / (SS_total + MS_within)

    # Effekt interpretasiyası (Cohen, 1988)
    def interpret_eta(eta):
        if eta < 0.01:
            return "Çox kiçik effekt"
        elif eta < 0.06:
            return "Kiçik effekt"
        elif eta < 0.14:
            return "Orta effekt"
        else:
            return "Böyük effekt"

    # --- Tukey HSD ---
    tukey = pairwise_tukeyhsd(endog=df[score_col],
                              groups=df[group_col],
                              alpha=0.05)

    return {
        "F-statistic": f_stat,
        "p-value": f"{p_val:.5f}",
        "Eta_squared (η²)": eta_sq,
        "Omega_squared (ω²)": omega_sq,
        "Effect_size_interpretation": interpret_eta(eta_sq),
        "Tukey_HSD_results": tukey.summary().as_text()
    }


# --- Nümunə istifadə ---

result = anova_effect_sizes_with_tukey(df, "Department", "Salary")

for k, v in result.items():
    print(f"{k}:\n{v}\n")
```

### Cohen's d, Hedges' g

```python
import numpy as np
import pandas as pd
from scipy import stats

def compare_groups(df, group_col, score_col):
    """
    İki qrup üçün:
    - Two-sample t-test
    - Cohen's d
    - Hedges' g
    - Effekt ölçüsü interpretasiyası
    
    Parametrlər:
        df (pd.DataFrame): Data
        group_col (str): Qrup sütunu
        score_col (str): Skor sütunu
    """
    # Qrupları ayırırıq
    groups = df[group_col].unique()
    if len(groups) != 2:
        raise ValueError("Funksiya yalnız iki qrup üçün işləyir.")
    
    group1 = df[df[group_col] == groups[0]][score_col].values
    group2 = df[df[group_col] == groups[1]][score_col].values

    # Two-sample t-test (equal variance fərziyyəsi ilə)
    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=True)

    # Ortalamalar və standart sapmalar
    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
    n1, n2 = len(group1), len(group2)

    # Pooled standart sapma
    s_pooled = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))

    # Cohen's d
    cohen_d = (mean1 - mean2) / s_pooled

    # Hedges' g (Cohen's d düzəlişli versiyası)
    correction_factor = 1 - (3 / (4*(n1+n2) - 9))
    hedges_g = cohen_d * correction_factor

    # Effekt ölçüsü interpretasiyası
    def interpret_effect(d):
        if abs(d) < 0.2:
            return "Kiçik effekt"
        elif abs(d) < 0.5:
            return "Orta effekt"
        elif abs(d) < 0.8:
            return "Böyük effekt"
        else:
            return "Çox böyük effekt"

    return {
        "t-statistic": t_stat,
        "p-value": f"{p_val:.5f}",
        "mean_group1": mean1,
        "mean_group2": mean2,
        "Cohen_d": cohen_d,
        "Hedges_g": hedges_g,
        "Effect_size_interpretation": interpret_effect(cohen_d)
    }

# --- Nümunə istifadə ---


result = compare_groups(df, "Gender", "Salary")

for k, v in result.items():
    print(f"{k}: {v}")
```

### Eta-squared (η²) və Omega-squared (ω²) və Tukey

```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def anova_effect_sizes_with_tukey(df, group_col, score_col):
    """
    Bir neçə qrup üçün:
    - ANOVA
    - Eta-squared (η²)
    - Omega-squared (ω²)
    - Tukey HSD post-hoc testi
    """
    # Qrupları ayırırıq
    groups = [df[df[group_col] == g][score_col].values for g in df[group_col].unique()]

    # --- ANOVA test ---
    f_stat, p_val = stats.f_oneway(*groups)
    
    # Bütün datanı birləşdiririk
    all_data = df[score_col].values
    grand_mean = np.mean(all_data)
    N = len(all_data)
    k = len(groups)

    # --- SS_total ---
    SS_total = np.sum((all_data - grand_mean) ** 2)

    # --- SS_between ---
    ns = [len(g) for g in groups]
    means = [np.mean(g) for g in groups]
    SS_between = sum([n * (m - grand_mean) ** 2 for n, m in zip(ns, means)])

    # --- SS_within ---
    SS_within = SS_total - SS_between

    # --- Azadlıq dərəcələri ---
    df_between = k - 1
    df_within = N - k

    # --- MS_within ---
    MS_within = SS_within / df_within

    # --- Effekt ölçüləri ---
    eta_sq = SS_between / SS_total
    omega_sq = (SS_between - (df_between * MS_within)) / (SS_total + MS_within)

    # Effekt interpretasiyası (Cohen, 1988)
    def interpret_eta(eta):
        if eta < 0.01:
            return "Çox kiçik effekt"
        elif eta < 0.06:
            return "Kiçik effekt"
        elif eta < 0.14:
            return "Orta effekt"
        else:
            return "Böyük effekt"

    # --- Tukey HSD ---
    tukey = pairwise_tukeyhsd(endog=df[score_col],
                              groups=df[group_col],
                              alpha=0.05)

    return {
        "F-statistic": f_stat,
        "p-value": f"{p_val:.5f}",
        "Eta_squared (η²)": eta_sq,
        "Omega_squared (ω²)": omega_sq,
        "Effect_size_interpretation": interpret_eta(eta_sq),
        "Tukey_HSD_results": tukey.summary().as_text()
    }


# --- Nümunə istifadə ---

result = anova_effect_sizes_with_tukey(df, "Department", "PerformanceScore")

for k, v in result.items():
    print(f"{k}:\n{v}\n")
```

```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def anova_effect_sizes_with_tukey(df, group_col, score_col):
    """
    Bir neçə qrup üçün:
    - ANOVA
    - Eta-squared (η²)
    - Omega-squared (ω²)
    - Tukey HSD post-hoc testi
    """
    # Qrupları ayırırıq
    groups = [df[df[group_col] == g][score_col].values for g in df[group_col].unique()]

    # --- ANOVA test ---
    f_stat, p_val = stats.f_oneway(*groups)
    
    # Bütün datanı birləşdiririk
    all_data = df[score_col].values
    grand_mean = np.mean(all_data)
    N = len(all_data)
    k = len(groups)

    # --- SS_total ---
    SS_total = np.sum((all_data - grand_mean) ** 2)

    # --- SS_between ---
    ns = [len(g) for g in groups]
    means = [np.mean(g) for g in groups]
    SS_between = sum([n * (m - grand_mean) ** 2 for n, m in zip(ns, means)])

    # --- SS_within ---
    SS_within = SS_total - SS_between

    # --- Azadlıq dərəcələri ---
    df_between = k - 1
    df_within = N - k

    # --- MS_within ---
    MS_within = SS_within / df_within

    # --- Effekt ölçüləri ---
    eta_sq = SS_between / SS_total
    omega_sq = (SS_between - (df_between * MS_within)) / (SS_total + MS_within)

    # Effekt interpretasiyası (Cohen, 1988)
    def interpret_eta(eta):
        if eta < 0.01:
            return "Çox kiçik effekt"
        elif eta < 0.06:
            return "Kiçik effekt"
        elif eta < 0.14:
            return "Orta effekt"
        else:
            return "Böyük effekt"

    # --- Tukey HSD ---
    tukey = pairwise_tukeyhsd(endog=df[score_col],
                              groups=df[group_col],
                              alpha=0.05)

    return {
        "F-statistic": f_stat,
        "p-value": f"{p_val:.5f}",
        "Eta_squared (η²)": eta_sq,
        "Omega_squared (ω²)": omega_sq,
        "Effect_size_interpretation": interpret_eta(eta_sq),
        "Tukey_HSD_results": tukey.summary().as_text()
    }


# --- Nümunə istifadə ---

result = anova_effect_sizes_with_tukey(df, "Shift", "AbsenceDays")

for k, v in result.items():
    print(f"{k}:\n{v}\n")
```

### Cramér’s V və φ (phi coefficient)

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def chi_square_with_effects(df, col1, col2):
    """
    İki kateqorik dəyişən üçün:
    - Chi-Square test
    - φ (phi coefficient) [2x2 tablolar üçün]
    - Cramér’s V (ümumiləşmiş phi)
    - Effekt ölçüsü interpretasiyası
    """
    # Kontingensiya cədvəli
    contingency_table = pd.crosstab(df[col1], df[col2])
    
    # Chi-Square testi
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    # Nümunə sayı
    n = contingency_table.sum().sum()
    
    # Effekt ölçüləri
    phi = np.sqrt(chi2 / n) if contingency_table.shape == (2, 2) else None
    k, r = contingency_table.shape[1], contingency_table.shape[0]
    cramers_v = np.sqrt(chi2 / (n * (min(r, k) - 1)))
    
    # Effekt interpretasiyası (Cohen, 1988)
    def interpret_cramers_v(v):
        if v < 0.1:
            return "Çox kiçik effekt"
        elif v < 0.3:
            return "Kiçik effekt"
        elif v < 0.5:
            return "Orta effekt"
        else:
            return "Böyük effekt"
    
    return {
        "Chi2-statistic": chi2,
        "p-value": f"{p:.5f}",
        "dof": dof,
        "phi (2x2)": phi if phi is not None else "Yalnız 2x2 tablolar üçün",
        "Cramér's V": cramers_v,
        "Effect_size_interpretation": interpret_cramers_v(cramers_v),
        "Contingency_table": contingency_table.to_string(),
        "Expected_table": pd.DataFrame(expected, 
                                       index=contingency_table.index, 
                                       columns=contingency_table.columns).to_string()
    }


# --- Nümunə istifadə ---

result = chi_square_with_effects(df, 'Region', 'Department')

for k, v in result.items():
    print(f"{k}:\n{v}\n")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# φ (phi) funksiyası (yalnız 2x2 cədvəl üçün)
def phi_coefficient(table):
    chi2, _, _, _ = chi2_contingency(table)
    n = table.sum().sum()
    return np.sqrt(chi2 / n)

# Cramér’s V funksiyası (ümumi mxn cədvəllər üçün)
def cramers_v(table):
    chi2, _, _, _ = chi2_contingency(table)
    n = table.sum().sum()
    r, k = table.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# Vizual funksiya
def plot_chi_effects(data, col1, col2):
    # Kontingensiya cədvəli
    table = pd.crosstab(data[col1], data[col2])
    chi2, p, _, expected = chi2_contingency(table)

    # Effekt ölçüləri
    phi_val = None
    if table.shape == (2, 2):
        phi_val = phi_coefficient(table)
    v_val = cramers_v(table)

    # Expected və residuals
    expected_table = pd.DataFrame(expected, 
                                  index=table.index, 
                                  columns=table.columns)
    residuals = (table - expected_table) / np.sqrt(expected_table)

    # --- Plot 1: Contingency Table ---
    plt.figure(figsize=(6, 5))
    sns.heatmap(table, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"{col1} × {col2} (Contingency Table)", fontsize=14)
    plt.tight_layout()
    plt.show()

    # --- Plot 2: Expected Table ---
    plt.figure(figsize=(6, 5))
    sns.heatmap(expected_table, annot=True, fmt=".1f", cmap="Greens", cbar=False)
    plt.title(f"{col1} × {col2} (Expected Table)", fontsize=14)
    plt.tight_layout()
    plt.show()

    # --- Plot 3: Residuals ---
    plt.figure(figsize=(6, 5))
    cmap = sns.diverging_palette(240, 10, as_cmap=True)  # mavi - qırmızı rəng palitrası
    sns.heatmap(residuals, annot=True, fmt=".2f", center=0, cmap=cmap, cbar=True)
    plt.title(f"{col1} × {col2} (Standardized Residuals)", fontsize=14)
    plt.tight_layout()
    plt.show()

    # --- Test nəticələri (konsola çıxarış) ---
    annotation = f"Chi² = {chi2:.2f}, p = {p:.4f}\n"
    if phi_val is not None:
        annotation += f"φ = {phi_val:.3f}\n"
    annotation += f"Cramér’s V = {v_val:.3f}"

    print(annotation)


# ----------------- TEST -----------------
# Nümunə data

plot_chi_effects(df, "SourceOfHire", "OfferAccepted")
```
