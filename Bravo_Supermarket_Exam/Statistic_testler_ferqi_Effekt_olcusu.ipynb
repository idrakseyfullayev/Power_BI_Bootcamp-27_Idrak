### **6. Python nÃ¼munÉ™si**

```python
import numpy as np

# Ä°ki qrup nÃ¼munÉ™si
group1 = np.array([12, 14, 15, 16, 14, 13])
group2 = np.array([10, 9, 12, 11, 10, 8])

# Ortalamalar
mean1, mean2 = np.mean(group1), np.mean(group2)

# Standart sapmalar
std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)

# NÃ¼munÉ™ Ã¶lÃ§Ã¼lÉ™ri
n1, n2 = len(group1), len(group2)

# Pooled standart sapma
s_pooled = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))

# Cohen's d
cohen_d = (mean1 - mean2) / s_pooled

# Hedges' g
hedges_g = cohen_d * (1 - (3 / (4*(n1+n2) - 9)))

print("Cohen's d:", cohen_d)
print("Hedges' g:", hedges_g)
```



## Cohenâ€™s d vÉ™ Hedgesâ€™ g Python code

```python
import numpy as np
from scipy import stats

def compare_groups(group1, group2):
    # Two-sample t-test (equal variance fÉ™rziyyÉ™si ilÉ™)
    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=True)

    # Ortalamalar vÉ™ standart sapmalar
    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
    n1, n2 = len(group1), len(group2)

    # Pooled standart sapma
    s_pooled = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))

    # Cohen's d
    cohen_d = (mean1 - mean2) / s_pooled

    # Hedges' g (Cohen's d dÃ¼zÉ™liÅŸli versiyasÄ±)
    correction_factor = 1 - (3 / (4*(n1+n2) - 9))
    hedges_g = cohen_d * correction_factor

    # Effekt Ã¶lÃ§Ã¼sÃ¼ interpretasiyasÄ±
    def interpret_effect(d):
        if abs(d) < 0.2:
            return "KiÃ§ik effekt"
        elif abs(d) < 0.5:
            return "Orta effekt"
        elif abs(d) < 0.8:
            return "BÃ¶yÃ¼k effekt"
        else:
            return "Ã‡ox bÃ¶yÃ¼k effekt"

    return {
        "t-statistic": t_stat,
        "p-value": p_val,
        "mean_group1": mean1,
        "mean_group2": mean2,
        "Cohen_d": cohen_d,
        "Hedges_g": hedges_g,
        "Effect_size_interpretation": interpret_effect(cohen_d)
    }

# NÃ¼munÉ™ istifadÉ™
group_a = [12, 14, 15, 16, 14, 13]
group_b = [10, 9, 12, 11, 10, 8]

result = compare_groups(group_a, group_b)
for k, v in result.items():
    print(f"{k}: {v}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def compare_groups_with_plot(group1, group2, group1_name="Group 1", group2_name="Group 2"):
    # Two-sample t-test (equal variance fÉ™rziyyÉ™si ilÉ™)
    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=True)

    # Ortalamalar vÉ™ standart sapmalar
    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
    n1, n2 = len(group1), len(group2)

    # Pooled standart sapma
    s_pooled = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))

    # Cohen's d
    cohen_d = (mean1 - mean2) / s_pooled

    # Hedges' g
    correction_factor = 1 - (3 / (4*(n1+n2) - 9))
    hedges_g = cohen_d * correction_factor

    # Effekt Ã¶lÃ§Ã¼sÃ¼ interpretasiyasÄ±
    def interpret_effect(d):
        if abs(d) < 0.2:
            return "KiÃ§ik effekt"
        elif abs(d) < 0.5:
            return "Orta effekt"
        elif abs(d) < 0.8:
            return "BÃ¶yÃ¼k effekt"
        else:
            return "Ã‡ox bÃ¶yÃ¼k effekt"

    interpretation = interpret_effect(cohen_d)

    # --- VizuallaÅŸdÄ±rma ---
    plt.figure(figsize=(7, 5))
    sns.boxplot(data=[group1, group2], palette="pastel")
    sns.stripplot(data=[group1, group2], color="black", alpha=0.5, jitter=True)

    # Annotasiya
    annotation_text = f"p-value = {p_val:.4f}\nCohen's d = {cohen_d:.3f} ({interpretation})"
    plt.text(0.5, max(max(group1), max(group2)) * 1.05,
             annotation_text, ha='center', fontsize=10, fontweight="bold")

    plt.xticks([0, 1], [group1_name, group2_name])
    plt.title("Group Comparison", fontsize=14, fontweight="bold")
    plt.ylabel("Values")
    plt.tight_layout()
    plt.show()

    # NÉ™ticÉ™lÉ™ri dict kimi qaytar
    return {
        "t-statistic": t_stat,
        "p-value": p_val,
        "mean_group1": mean1,
        "mean_group2": mean2,
        "Cohen_d": cohen_d,
        "Hedges_g": hedges_g,
        "Effect_size_interpretation": interpretation
    }

# NÃ¼munÉ™ istifadÉ™
group_a = [12, 14, 15, 16, 14, 13]
group_b = [10, 9, 12, 11, 10, 8]

result = compare_groups_with_plot(group_a, group_b, group1_name="Filial A", group2_name="Filial B")
for k, v in result.items():
    print(f"{k}: {v}")
```
