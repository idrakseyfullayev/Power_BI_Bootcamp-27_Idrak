## Cohenâ€™s d, Hedgesâ€™ g

Bunlar hÉ™r ikisi **effect size** (fÉ™rqin bÃ¶yÃ¼klÃ¼yÃ¼) Ã¶lÃ§Ã¼lÉ™ridir vÉ™ **Two-sample t-test**-dÉ™ istifadÉ™ olunur ki, fÉ™rqin **praktik É™hÉ™miyyÉ™tini** (real hÉ™yatda nÉ™ qÉ™dÉ™r Ã¶nÉ™mli olduÄŸunu) deyÉ™ bilÉ™k.

### **1. NiyÉ™ lazÄ±m olur?**

Two-sample t-test sÉ™nÉ™ yalnÄ±z **â€œfÉ™rq var, ya yoxâ€** (p-value) deyir.
Amma **p-value â‰  fÉ™rqin bÃ¶yÃ¼klÃ¼yÃ¼**.

* BÃ¶yÃ¼k nÃ¼munÉ™ gÃ¶tÃ¼rsÉ™n, kiÃ§ik fÉ™rq belÉ™ p<0.05 verÉ™ bilÉ™r.
* KiÃ§ik nÃ¼munÉ™dÉ™ isÉ™ bÃ¶yÃ¼k fÉ™rq belÉ™ bÉ™zÉ™n statistik É™hÉ™miyyÉ™t daÅŸÄ±maya bilÉ™r.

Cohenâ€™s d vÉ™ Hedgesâ€™ g bu problemi hÉ™ll edir, Ã§Ã¼nki onlar **standart sapma ilÉ™ Ã¶lÃ§Ã¼lÉ™n normallaÅŸdÄ±rÄ±lmÄ±ÅŸ fÉ™rq** verirlÉ™r.

---

### **2. Cohenâ€™s d â€“ riyazi tÉ™rifi**

Ä°ki qrup Ã¼Ã§Ã¼n Cohenâ€™s d:

$$
d = \frac{\bar{X}_1 - \bar{X}_2}{s_{pooled}}
$$

Burada:

* $\bar{X}_1, \bar{X}_2$ â†’ iki qrupun ortalamalarÄ±
* $s_{pooled}$ â†’ **birgÉ™ (pooled) standart sapma**

BirgÉ™ standart sapma formulu:

$$
s_{pooled} = \sqrt{\frac{(n_1 - 1)s_1^2 + (n_2 - 1)s_2^2}{n_1 + n_2 - 2}}
$$

* $n_1, n_2$ â†’ nÃ¼munÉ™ saylarÄ±
* $s_1, s_2$ â†’ qruplarÄ±n standart sapmalarÄ±

---

### **Cohenâ€™s d interpretasiyasÄ±** (Cohen, 1988):

* **0.2** â†’ kiÃ§ik effekt
* **0.5** â†’ orta effekt
* **0.8** â†’ bÃ¶yÃ¼k effekt

MÉ™sÉ™lÉ™n, d = 0.8 demÉ™kdir ki, qruplarÄ±n ortalamalarÄ± 0.8 standart sapma qÉ™dÉ™r fÉ™rqlÉ™nir (bÃ¶yÃ¼k fÉ™rqdir).

---

### **3. Hedgesâ€™ g â€“ niyÉ™ var?**

Cohenâ€™s d kiÃ§ik nÃ¼munÉ™lÉ™rdÉ™ bir az **yuxarÄ± ÅŸiÅŸirdilmiÅŸ** ola bilÉ™r.
Hedgesâ€™ g bu qÉ™rÉ™zi dÃ¼zÉ™ldir.

Formulu:

$$
g = d \times \left(1 - \frac{3}{4(n_1 + n_2) - 9}\right)
$$

Burada:

* $d$ â†’ Cohenâ€™s d
* DÃ¼zÉ™liÅŸ É™msalÄ± â†’ $1 - \frac{3}{4N - 9}$, burada $N = n_1 + n_2$

âž¡ YÉ™ni Hedgesâ€™ g â‰ˆ Cohenâ€™s d, amma **kiÃ§ik nÃ¼munÉ™lÉ™rdÉ™ daha dÉ™qiq**.

---

### **4. HansÄ±nda istifadÉ™ olunur?**

* BÃ¶yÃ¼k nÃ¼munÉ™lÉ™r â†’ Cohenâ€™s d kifayÉ™tdir.
* KiÃ§ik nÃ¼munÉ™lÉ™r (mÉ™sÉ™lÉ™n, hÉ™r qrup < 20) â†’ Hedgesâ€™ g daha doÄŸru nÉ™ticÉ™ verir.

---

### **5. NiyÉ™ Two-sample t-test ilÉ™ birlikdÉ™ iÅŸlÉ™yir?**

* t-test p-value ilÉ™ **â€œstatistik É™hÉ™miyyÉ™tâ€** deyir.
* Cohenâ€™s d / Hedgesâ€™ g isÉ™ **â€œpraktik É™hÉ™miyyÉ™tâ€** deyir.
* BirlikdÉ™: â€œFÉ™rq varâ€ + â€œfÉ™rq bÃ¶yÃ¼kdÃ¼r/orta/kiÃ§ikdirâ€ kimi nÉ™ticÉ™ verilir.


## Cohen vÉ™ Hedges Ã¼Ã§Ã¼n effekt Ã¶lÃ§Ã¼sÃ¼ cÉ™dvÉ™li


Cohenâ€™s *d* vÉ™ Hedgesâ€™ *g* ikili qruplarÄ±n (iki ortalamanÄ±n) mÃ¼qayisÉ™sindÉ™ **effekt Ã¶lÃ§Ã¼sÃ¼nÃ¼** gÃ¶stÉ™rmÉ™k Ã¼Ã§Ã¼n istifadÉ™ olunur.

ðŸ“Œ Æsas fÉ™rq:

* **Cohenâ€™s d** â€“ sadÉ™ standartlaÅŸdÄ±rÄ±lmÄ±ÅŸ fÉ™rqdir.
* **Hedgesâ€™ g** â€“ Cohenâ€™s d-nin *kiÃ§ik nÃ¼munÉ™lÉ™rdÉ™* (small sample bias) dÃ¼zÉ™liÅŸ olunmuÅŸ versiyasÄ±dÄ±r.

---

### ðŸ“Š Effekt Ã¶lÃ§Ã¼lÉ™ri Ã¼Ã§Ã¼n tÉ™xmini interpretasiya (Cohen, 1988; Hedges, 1981)

| Effekt Ã¶lÃ§Ã¼sÃ¼ | KiÃ§ik (small) | Orta (medium) | BÃ¶yÃ¼k (large) |
| ------------- | ------------- | ------------- | ------------- |
| **Cohenâ€™s d** | 0.20 â€“ 0.49   | 0.50 â€“ 0.79   | â‰¥ 0.80        |
| **Hedgesâ€™ g** | 0.20 â€“ 0.49   | 0.50 â€“ 0.79   | â‰¥ 0.80        |

---

âœ… ÆgÉ™r Cohenâ€™s *d* vÉ™ ya Hedgesâ€™ *g* **0.2 â€“ 0.49** civarÄ±nda Ã§Ä±xÄ±rsa â†’ zÉ™if effekt
âœ… **0.50 â€“ 0.79** â†’ orta effekt
âœ… **0.8 vÉ™ yuxarÄ±** â†’ gÃ¼clÃ¼ effekt

âš ï¸ Qeyd: Cohenâ€™s d vÉ™ Hedgesâ€™ g Ã§ox vaxt **t-test** nÉ™ticÉ™lÉ™rinÉ™ paralel olaraq istifadÉ™ olunur. YÉ™ni `p-value` deyir ki, fÉ™rq statistik olaraq É™hÉ™miyyÉ™tlidir ya yox, effekt Ã¶lÃ§Ã¼sÃ¼ isÉ™ fÉ™rqin **praktik olaraq gÃ¼clÃ¼lÃ¼yÃ¼nÃ¼** gÃ¶stÉ™rir.



## Cohenâ€™s d vÉ™ Hedgesâ€™ g Python code

* Ä°ki qrup Ã¼Ã§Ã¼n **Two-sample t-test** edÉ™cÉ™k
* **p-value** verÉ™cÉ™k
* **Cohenâ€™s d** vÉ™ **Hedgesâ€™ g** hesablayacaq
* Effekt Ã¶lÃ§Ã¼sÃ¼nÃ¼n **interpretasiyasÄ±nÄ±** (kiÃ§ik/orta/bÃ¶yÃ¼k) Ã§Ä±xaracaq



```python
import numpy as np
import pandas as pd
from scipy import stats

def compare_groups(df, group_col, score_col):
    """
    Ä°ki qrup Ã¼Ã§Ã¼n:
    - Two-sample t-test
    - Cohen's d
    - Hedges' g
    - Effekt Ã¶lÃ§Ã¼sÃ¼ interpretasiyasÄ±
    
    ParametrlÉ™r:
        df (pd.DataFrame): Data
        group_col (str): Qrup sÃ¼tunu
        score_col (str): Skor sÃ¼tunu
    """
    # QruplarÄ± ayÄ±rÄ±rÄ±q
    groups = df[group_col].unique()
    if len(groups) != 2:
        raise ValueError("Funksiya yalnÄ±z iki qrup Ã¼Ã§Ã¼n iÅŸlÉ™yir.")
    
    group1 = df[df[group_col] == groups[0]][score_col].values
    group2 = df[df[group_col] == groups[1]][score_col].values

    # Two-sample t-test (equal variance fÉ™rziyyÉ™si ilÉ™)
    t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=True)

    # Ortalamalar vÉ™ standart sapmalar
    mean1, mean2 = np.mean(group1), np.mean(group2)
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
    n1, n2 = len(group1), len(group2)

    # Pooled standart sapma
    s_pooled = np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))

    # Cohen's d
    cohen_d = (mean1 - mean2) / s_pooled

    # Hedges' g (Cohen's d dÃ¼zÉ™liÅŸli versiyasÄ±)
    correction_factor = 1 - (3 / (4*(n1+n2) - 9))
    hedges_g = cohen_d * correction_factor

    # Effekt Ã¶lÃ§Ã¼sÃ¼ interpretasiyasÄ±
    def interpret_effect(d):
        if abs(d) < 0.2:
            return "KiÃ§ik effekt"
        elif abs(d) < 0.5:
            return "Orta effekt"
        elif abs(d) < 0.8:
            return "BÃ¶yÃ¼k effekt"
        else:
            return "Ã‡ox bÃ¶yÃ¼k effekt"

    return {
        "t-statistic": t_stat,
        "p-value": f"{p_val:.5f}",
        "mean_group1": mean1,
        "mean_group2": mean2,
        "Cohen_d": cohen_d,
        "Hedges_g": hedges_g,
        "Effect_size_interpretation": interpret_effect(cohen_d)
    }

# --- NÃ¼munÉ™ istifadÉ™ ---
data = {
    "Group": ["A","A","A","A","A","A",
              "B","B","B","B","B","B"],
    "Score": [12,14,15,16,14,13,
              10, 9,12,11,10, 8]
}

df = pd.DataFrame(data)

result = compare_groups(df, "Group", "Score")

for k, v in result.items():
    print(f"{k}: {v}")
```


## Eta-squared (Î·Â²) vÉ™ Omega-squared (Ï‰Â²)



### ðŸ”¹ 1. Eta-squared (Î·Â²)

**TÉ™rif:**
Eta-squared Ã¼mumi dispersiyanÄ±n (variansÄ±n) neÃ§É™ faizinin qruplar arasÄ±ndakÄ± fÉ™rqlÉ™ izah olunduÄŸunu gÃ¶stÉ™rir.

**Riyazi dÃ¼stur:**

$$
\eta^2 = \frac{SS_{between}}{SS_{total}}
$$

Burada:

* $SS_{between} = \sum_{i=1}^k n_i \cdot (\bar{X}_i - \bar{X})^2$
* $SS_{total} = \sum_{i=1}^k \sum_{j=1}^{n_i} (X_{ij} - \bar{X})^2$

---

### ðŸ”¹ 2. Omega-squared (Ï‰Â²)

**Riyazi dÃ¼stur:**

$$
\omega^2 = \frac{SS_{between} - (df_{between} \cdot MS_{within})}{SS_{total} + MS_{within}}
$$

Burada:

* $SS_{between} = \sum_{i=1}^k n_i \cdot (\bar{X}_i - \bar{X})^2$
* $SS_{total} = \sum_{i=1}^k \sum_{j=1}^{n_i} (X_{ij} - \bar{X})^2$
* $df_{between} = k - 1$  (k = qrup sayÄ±)
* $SS_{within} = \sum_{i=1}^k \sum_{j=1}^{n_i} (X_{ij} - \bar{X}_i)^2$
* $df_{within} = N - k$  (N = bÃ¼tÃ¼n mÃ¼ÅŸahidÉ™lÉ™rin sayÄ±)
* $MS_{within} = \frac{SS_{within}}{df_{within}}$

---

### ðŸ”¹ 3. NiyÉ™ ANOVA ilÉ™ birlikdÉ™ istifadÉ™ olunur?

* **ANOVA â†’ p-value**: qruplar arasÄ±nda fÉ™rq varmÄ±?
* **Î·Â² vÉ™ ya Ï‰Â² â†’ effekt Ã¶lÃ§Ã¼sÃ¼**: fÉ™rq nÉ™ qÉ™dÉ™r gÃ¼clÃ¼dÃ¼r?

MÉ™sÉ™lÉ™n:

* p-value < 0.05 (statistik fÉ™rq var)
* Î·Â² = 0.15 â†’ bu o demÉ™kdir ki, Ã¼mumi dispersiyanÄ±n **15%-i qruplar arasÄ±ndakÄ± fÉ™rqlÉ™ izah olunur** â†’ bÃ¶yÃ¼k effekt.

## Eta-squared (Î·Â²) vÉ™ Omega-squared (Ï‰Â²) vÉ™ Tukey Python Code



### ðŸ“Œ Kod

```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

def anova_effect_sizes_with_tukey(df, group_col, score_col):
    """
    Bir neÃ§É™ qrup Ã¼Ã§Ã¼n:
    - ANOVA
    - Eta-squared (Î·Â²)
    - Omega-squared (Ï‰Â²)
    - Tukey HSD post-hoc testi
    """
    # QruplarÄ± ayÄ±rÄ±rÄ±q
    groups = [df[df[group_col] == g][score_col].values for g in df[group_col].unique()]

    # --- ANOVA test ---
    f_stat, p_val = stats.f_oneway(*groups)
    
    # BÃ¼tÃ¼n datanÄ± birlÉ™ÅŸdiririk
    all_data = df[score_col].values
    grand_mean = np.mean(all_data)
    N = len(all_data)
    k = len(groups)

    # --- SS_total ---
    SS_total = np.sum((all_data - grand_mean) ** 2)

    # --- SS_between ---
    ns = [len(g) for g in groups]
    means = [np.mean(g) for g in groups]
    SS_between = sum([n * (m - grand_mean) ** 2 for n, m in zip(ns, means)])

    # --- SS_within ---
    SS_within = SS_total - SS_between

    # --- AzadlÄ±q dÉ™rÉ™cÉ™lÉ™ri ---
    df_between = k - 1
    df_within = N - k

    # --- MS_within ---
    MS_within = SS_within / df_within

    # --- Effekt Ã¶lÃ§Ã¼lÉ™ri ---
    eta_sq = SS_between / SS_total
    omega_sq = (SS_between - (df_between * MS_within)) / (SS_total + MS_within)

    # Effekt interpretasiyasÄ± (Cohen, 1988)
    def interpret_eta(eta):
        if eta < 0.01:
            return "Ã‡ox kiÃ§ik effekt"
        elif eta < 0.06:
            return "KiÃ§ik effekt"
        elif eta < 0.14:
            return "Orta effekt"
        else:
            return "BÃ¶yÃ¼k effekt"

    # --- Tukey HSD ---
    tukey = pairwise_tukeyhsd(endog=df[score_col],
                              groups=df[group_col],
                              alpha=0.05)

    return {
        "F-statistic": f_stat,
        "p-value": f"{p_val:.5f}",
        "Eta_squared (Î·Â²)": eta_sq,
        "Omega_squared (Ï‰Â²)": omega_sq,
        "Effect_size_interpretation": interpret_eta(eta_sq),
        "Tukey_HSD_results": tukey.summary().as_text()
    }


# --- NÃ¼munÉ™ istifadÉ™ ---
data = {
    "Group": ["A","A","A","A","A","A",
              "B","B","B","B","B","B",
              "C","C","C","C","C","C"],
    "Score": [12,14,15,16,14,13,
              10, 9,12,11,10, 8,
              18,20,19,17,21,20]
}

df = pd.DataFrame(data)

result = anova_effect_sizes_with_tukey(df, "Group", "Score")

for k, v in result.items():
    print(f"{k}:\n{v}\n")
```

### ðŸ“Š Bu funksiya nÉ™ edir:

* **ANOVA** aparÄ±r â†’ F-statistic vÉ™ p-value Ã§Ä±xarÄ±r.
* **Î·Â² vÉ™ Ï‰Â²** effekt Ã¶lÃ§Ã¼lÉ™rini hesablayÄ±r.
* Î·Â² Ã¼Ã§Ã¼n interpretasiya (kiÃ§ik, orta, bÃ¶yÃ¼k) verir.
* Tukey HSD â†’ HÉ™r bir qrup cÃ¼tÃ¼ Ã¼Ã§Ã¼n post-hoc mÃ¼qayisÉ™ nÉ™ticÉ™s
* Ä°stÉ™nilÉ™n sayda qrup qÉ™bul edir (`anova_effect_sizes(group1, group2, group3, ...)`).


---

Bu, **Tukey HSD post-hoc testinin nÉ™ticÉ™ cÉ™dvÉ™lidir** vÉ™ hÉ™r sÃ¼tunun mÉ™nasÄ± belÉ™dir:

| SÃ¼tun        | Ä°zahÄ±                                                                                                          |
| ------------ | -------------------------------------------------------------------------------------------------------------- |
| **group1**   | MÃ¼qayisÉ™ olunan birinci qrup                                                                                   |
| **group2**   | MÃ¼qayisÉ™ olunan ikinci qrup                                                                                    |
| **meandiff** | QruplarÄ±n **ortalama fÉ™rqi** (group1 â€“ group2)                                                                 |
| **p-adj**    | Tukey HSD ilÉ™ hesablanmÄ±ÅŸ **dÃ¼zÉ™ldilmiÅŸ p-value** (bir neÃ§É™ mÃ¼qayisÉ™yÉ™ uyÄŸun)                                  |

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Î·Â² funksiyasÄ± (istÉ™nilÉ™n sayda qrup Ã¼Ã§Ã¼n)
def eta_squared(*groups):
    all_data = np.concatenate(groups)
    grand_mean = np.mean(all_data)

    ss_total = np.sum((all_data - grand_mean) ** 2)
    ss_between = sum(len(g) * (np.mean(g) - grand_mean) ** 2 for g in groups)

    eta2 = ss_between / ss_total
    return eta2

# Vizual versiya
def plot_anova_eta2(data, group_col, value_col):
    groups = [data[data[group_col] == g][value_col].values for g in data[group_col].unique()]
    f_stat, p_val = f_oneway(*groups)
    eta2 = eta_squared(*groups)

    # Boxplot
    plt.figure(figsize=(8, 6))
    ax = sns.boxplot(x=group_col, y=value_col, hue=group_col, data=data, palette="pastel")

    # Median annotasiyasÄ±
    for i, g in enumerate(data[group_col].unique()):
        median_val = np.median(data[data[group_col] == g][value_col])
        ax.text(i, median_val, f'{median_val:.2f}', horizontalalignment='center',
                color='black', weight='semibold')    
    
    
    # Annotasiya
    plt.text(0.8, data[value_col].max()*0.8,
             f"ANOVA: F = {f_stat:.2f}, p = {p_val:.4f}, Î·Â² = {eta2:.3f}",
             ha="center", va="bottom", fontsize=12, color="darkred")

    plt.title("Qruplar Ã¼zrÉ™ mÃ¼qayisÉ™ (ANOVA + Î·Â²)", fontsize=14)
    plt.tight_layout()
    plt.show()

# ----------------- TEST -----------------
# NÃ¼munÉ™ data (istÉ™nilÉ™n sayda qrup ola bilÉ™r)
np.random.seed(42)
df = pd.DataFrame({
    "Group": np.repeat(["A", "B", "C", "D"], 30),
    "Score": np.concatenate([
        np.random.normal(50, 10, 30),
        np.random.normal(55, 10, 30),
        np.random.normal(60, 10, 30),
        np.random.normal(65, 10, 30),
    ])
})

plot_anova_eta2(df, "Group", "Score")
```

## Chi-square test of independence Python code


```python
import pandas as pd
from scipy.stats import chi2_contingency

# NÃ¼munÉ™ data
data = {
    "Group1": ["North","North","North","South","South","South","East","East","East","West","West","West"],
    "Group2": ["HR","IT","Finance","HR","IT","Finance","HR","IT","Finance","HR","IT","Finance"]
}

df = pd.DataFrame(data)

# Contingency table (kÉ™siÅŸmÉ™ cÉ™dvÉ™li)
contingency_table = pd.crosstab(df['Group1'], df['Group2'])

# Chi-square testi
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square dÉ™yÉ™ri:", chi2)
print("p-value:", f"{p:.5f}")
print("AzadlÄ±q dÉ™rÉ™cÉ™si (dof):", dof)
print("Contingency_table:\n", contingency_table.values)
print("GÃ¶zlÉ™nilÉ™n dÉ™yÉ™rlÉ™r:\n", expected)

# NÉ™ticÉ™nin ÅŸÉ™rhi
if p < 0.05:
    print("NÉ™ticÉ™: Group1 ilÉ™ Group2 arasÄ±nda statistik É™hÉ™miyyÉ™tli É™laqÉ™ var.")
else:
    print("NÉ™ticÉ™: Group1 ilÉ™ Group2 arasÄ±nda statistik É™hÉ™miyyÉ™tli É™laqÉ™ yoxdur.")
```

V = \sqrt{\frac{\chi^2}{N \cdot (k-1)}}
$$

Burada:

* \$\chi^2\$ â†’ Chi-Square statistikasÄ±
* \$N\$ â†’ Ã¼mumi mÃ¼ÅŸahidÉ™ sayÄ±
* \$k = \min(r, c)\$ â†’ sÉ™tir vÉ™ sÃ¼tun saylarÄ±nÄ±n kiÃ§iyi

### **CramÃ©râ€™s V interpretasiyasÄ± (Cohen, 1988):**

* **0.1** â†’ zÉ™if É™laqÉ™
* **0.3** â†’ orta É™laqÉ™
* **0.5** â†’ gÃ¼clÃ¼ É™laqÉ™

MÉ™sÉ™lÉ™n, V = 0.3 Ã§Ä±xarsa, bu **orta gÃ¼clÃ¼ É™laqÉ™** demÉ™kdir.

### **5. NiyÉ™ Chi-Square testi ilÉ™ birlikdÉ™ iÅŸlÉ™yir?**

* HÉ™r ikisi **Chi-Square statistikasÄ±** É™sasÄ±nda hesablanÄ±r vÉ™ **kateqorik dÉ™yiÅŸÉ™nlÉ™r arasÄ±ndakÄ± É™laqÉ™nin gÃ¼cÃ¼nÃ¼** Ã¶lÃ§Ã¼r.




## CramÃ©râ€™s V vÉ™ Ï† (Phi) effekt vÉ™ Standardized_residuals Python code

### ðŸ“Œ Kod

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def chi_square_with_effects(df, col1, col2):
    """
    Ä°ki kateqorik dÉ™yiÅŸÉ™n Ã¼Ã§Ã¼n:
    - Chi-Square test
    - Ï† (phi coefficient) [2x2 tablolar Ã¼Ã§Ã¼n]
    - CramÃ©râ€™s V (Ã¼mumilÉ™ÅŸmiÅŸ phi)
    - Effekt Ã¶lÃ§Ã¼sÃ¼ interpretasiyasÄ±
    - StandartlaÅŸdÄ±rÄ±lmÄ±ÅŸ residuals
    """
    # Kontingensiya cÉ™dvÉ™li
    contingency_table = pd.crosstab(df[col1], df[col2])
    
    # Chi-Square testi
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    # NÃ¼munÉ™ sayÄ±
    n = contingency_table.sum().sum()
    
    # Effekt Ã¶lÃ§Ã¼lÉ™ri
    phi = np.sqrt(chi2 / n) if contingency_table.shape == (2, 2) else None
    k, r = contingency_table.shape[1], contingency_table.shape[0]
    cramers_v = np.sqrt(chi2 / (n * (min(r, k) - 1)))
    
    # Effekt interpretasiyasÄ± (Cohen, 1988)
    def interpret_cramers_v(v):
        if v < 0.1:
            return "Ã‡ox kiÃ§ik effekt"
        elif v < 0.3:
            return "KiÃ§ik effekt"
        elif v < 0.5:
            return "Orta effekt"
        else:
            return "BÃ¶yÃ¼k effekt"
    
    # Expected vÉ™ residuals
    expected_table = pd.DataFrame(expected, 
                                  index=contingency_table.index, 
                                  columns=contingency_table.columns)
    
    residuals = (contingency_table - expected_table) / np.sqrt(expected_table)
    
    return {
        "Chi2-statistic": chi2,
        "p-value": f"{p:.5f}",
        "dof": dof,
        "phi (2x2)": phi if phi is not None else "YalnÄ±z 2x2 tablolar Ã¼Ã§Ã¼n",
        "CramÃ©r's V": cramers_v,
        "Effect_size_interpretation": interpret_cramers_v(cramers_v),
        "Contingency_table": contingency_table.to_string(),
        "Expected_table": expected_table.to_string(),
        "Standardized_residuals": residuals.round(2).to_string()
    }


# --- NÃ¼munÉ™ istifadÉ™ ---
data = {
    "Gender": ["M","M","F","F","M","F","M","F","M","F","F","M"],
    "Preference": ["A","B","A","A","B","B","A","C","C","C","A","B"]
}

df = pd.DataFrame(data)

result = chi_square_with_effects(df, "Gender", "Preference")

for k, v in result.items():
    print(f"{k}:\n{v}\n")
```

* **Chi-Square test** â†’ $\chi^2, p$-value, dof
* **Ï† (phi coefficient)** â†’ yalnÄ±z 2x2 cÉ™dvÉ™llÉ™rdÉ™ Ã§Ä±xÄ±r
* **CramÃ©râ€™s V** â†’ Ã¼mumi effekt Ã¶lÃ§Ã¼sÃ¼ (hÉ™r Ã¶lÃ§Ã¼lÃ¼ kontingensiya cÉ™dvÉ™li Ã¼Ã§Ã¼n)
* **Interpretasiya** (Cohen 1988) ilÉ™ effekt dÉ™rÉ™cÉ™sini gÃ¶stÉ™rir
* **Residual cÉ™dvÉ™lini** hesablayÄ±r
* ÆlavÉ™ olaraq **gÃ¶zlÉ™nilÉ™n tezliklÉ™r cÉ™dvÉ™lini** dÉ™ Ã§ap edir




## Contingency Table & Expected Table & Standardized Residuals Heatmap


### ðŸ“Œ Kod

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Ï† (phi) funksiyasÄ± (yalnÄ±z 2x2 cÉ™dvÉ™l Ã¼Ã§Ã¼n)
def phi_coefficient(table):
    chi2, _, _, _ = chi2_contingency(table)
    n = table.sum().sum()
    return np.sqrt(chi2 / n)

# CramÃ©râ€™s V funksiyasÄ± (Ã¼mumi mxn cÉ™dvÉ™llÉ™r Ã¼Ã§Ã¼n)
def cramers_v(table):
    chi2, _, _, _ = chi2_contingency(table)
    n = table.sum().sum()
    r, k = table.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# Vizual funksiya
def plot_chi_effects(data, col1, col2):
    # Kontingensiya cÉ™dvÉ™li
    table = pd.crosstab(data[col1], data[col2])
    chi2, p, _, expected = chi2_contingency(table)

    # Effekt Ã¶lÃ§Ã¼lÉ™ri
    phi_val = None
    if table.shape == (2, 2):
        phi_val = phi_coefficient(table)
    v_val = cramers_v(table)

    # Expected vÉ™ residuals
    expected_table = pd.DataFrame(expected, 
                                  index=table.index, 
                                  columns=table.columns)
    residuals = (table - expected_table) / np.sqrt(expected_table)

    # --- Plot 1: Contingency Table ---
    plt.figure(figsize=(6, 5))
    sns.heatmap(table, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"{col1} Ã— {col2} (Contingency Table)", fontsize=14)
    plt.tight_layout()
    plt.show()

    # --- Plot 2: Expected Table ---
    plt.figure(figsize=(6, 5))
    sns.heatmap(expected_table, annot=True, fmt=".1f", cmap="Greens", cbar=False)
    plt.title(f"{col1} Ã— {col2} (Expected Table)", fontsize=14)
    plt.tight_layout()
    plt.show()

    # --- Plot 3: Residuals ---
    plt.figure(figsize=(6, 5))
    cmap = sns.diverging_palette(240, 10, as_cmap=True)  # mavi - qÄ±rmÄ±zÄ± rÉ™ng palitrasÄ±
    sns.heatmap(residuals, annot=True, fmt=".2f", center=0, cmap=cmap, cbar=True)
    plt.title(f"{col1} Ã— {col2} (Standardized Residuals)", fontsize=14)
    plt.tight_layout()
    plt.show()

    # --- Test nÉ™ticÉ™lÉ™ri (konsola Ã§Ä±xarÄ±ÅŸ) ---
    annotation = f"ChiÂ² = {chi2:.2f}, p = {p:.4f}\n"
    if phi_val is not None:
        annotation += f"Ï† = {phi_val:.3f}\n"
    annotation += f"CramÃ©râ€™s V = {v_val:.3f}"

    print(annotation)


# ----------------- TEST -----------------
# NÃ¼munÉ™ data
np.random.seed(42)
df = pd.DataFrame({
    "Gender": np.random.choice(["Male", "Female"], 200),
    "Preference": np.random.choice(["A", "B", "C"], 200, p=[0.3, 0.5, 0.2])
})

plot_chi_effects(df, "Gender", "Preference")
```
