## Agro West Analys



## 1️⃣ KPI nədir

**KPI (Key Performance Indicator)** – əsas performans göstəricisi, yəni bir işin və ya bölmənin uğurunu ölçmək üçün istifadə olunan konkret göstəricidir.
Məsələn: satış həcmi, müştəri məmnuniyyəti, anbar dövriyyəsi, çatdırılma vaxtı və s.

---

## 2️⃣ KPI-ların növləri Aqro West üçün

Aqro West-in fəaliyyəti əsasən **satış, logistika və anbar** ətrafında olduğundan, bəzi KPI-lar bunlar ola bilər:

**Satış üzrə KPI-lar:**

* Aylıq satış həcmi (AZN)
* Satış artım faizi (%)
* Top məhsul kateqoriyaları üzrə satış

**Logistika / Anbar KPI-ları:**

* Ortalama çatdırılma müddəti (günlə)
* Anbarda məhsul dövriyyəsi (inventory turnover)
* Ziyana gedən məhsul faizi (%)

**Müştəri xidməti KPI-ları:**

* Müştəri şikayətləri sayı
* Müştəri məmnuniyyəti reytinqi

---

## 3️⃣ Excel-də KPI quraşdırmaq

### 3.1 Addım: Data toplamaq

* Məhsulların adları
* Satış tarixi və məbləği
* Çatdırılma vaxtları
* Anbarda mövcudluq və ziyan məlumatları

**Nümunə CSV faylı:**

| Məhsul | Satış tarixi | Satış miqdarı | Çatdırılma günləri | Anbar miqdarı | Ziyan % |
| ------ | ------------ | ------------- | ------------------ | ------------- | ------- |
| Süd    | 01.09.2025   | 200           | 2                  | 500           | 1       |
| Çörək  | 01.09.2025   | 150           | 1                  | 300           | 0.5     |

---

### 3.2 Addım: Hesablama formulları

1. **Aylıq satış həcmi:**

```excel
=SUMIF(B:B, ">=01.09.2025", C:C) - SUMIF(B:B, ">30.09.2025", C:C)
```

və ya **Pivot Table** ilə ay üzrə toplamaq.

2. **Satış artım faizi:**
   Əgər əvvəlki ay satış həcmi `E2`, cari ay `F2`-dədirsə:

```excel
=(F2-E2)/E2
```

3. **Ortalama çatdırılma müddəti:**

```excel
=AVERAGE(D2:D1000)
```

4. **Ziyana gedən məhsul faizi:**

```excel
=AVERAGE(F2:F1000)
```

---

### 3.3 Addım: Dashboard yaratmaq

1. Excel-də **Pivot Table** və **Charts** istifadə edərək KPI-ları vizuallaşdırın.
2. Satış artımı üçün **line chart**, anbar dövriyyəsi üçün **bar chart**, ziyan üçün **gauge chart** istifadə etmək olar.

## 4️⃣ Nəqliyyat KPI-ları

**Əsas məqsəd:** Məhsulların vaxtında çatdırılması və nəqliyyat xərclərinin optimallaşdırılması.

### Nəqliyyat üzrə KPI-lar

1. **Ortalama çatdırılma müddəti (günlə)**

   * Hər çatdırılma üçün keçən günləri hesablamaq.
   * Formula Excel-də:

   ```excel
   =AVERAGE(Çatdırılma_Günləri_Sütunu)
   ```

2. **Çatdırılma vaxtına riayət faizi (%)**

   * Planlaşdırılan çatdırılma günləri ilə real günləri müqayisə etmək.
   * Excel-də:

   ```excel
   =COUNTIF(Çatdırılma_Günləri_Sütunu,"<="&Planlaşdırılmış_Gün_Sütunu)/COUNTA(Çatdırılma_Günləri_Sütunu)
   ```

3. **Nəqliyyat xərci (AZN)**

   * Hər göndəriş üçün xərcləri toplamaq.
   * Formula:

   ```excel
   =SUM(Nəqliyyat_Xərci_Sütunu)
   ```

4. **Sifariş başına orta nəqliyyat xərci (AZN)**

   * Xərcləri sifariş sayına bölmək.
   * Formula:

   ```excel
   =AVERAGE(Nəqliyyat_Xərci_Sütunu)
   ```

5. **Daşıma növü üzrə pay (%)**

   * Tır, Minivan, Avtomobil və s. paylarını hesablamaq.
   * Excel Pivot Table və ya COUNTIF ilə hesablana bilər:

   ```excel
   =COUNTIF(Nəqliyyat_Növü_Sütunu,"Tır")/COUNTA(Nəqliyyat_Növü_Sütunu)
   ```
